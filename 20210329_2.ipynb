```python
import pandas as pd

df = pd.read_csv('1st.csv')
df['Total_Sales'] = df['NA_Sales']+df['EU_Sales']+df['JP_Sales']+df['Other_Sales']

df = df.drop('Unnamed: 0', axis =1)
df_70 = df['Year'].isin(['1970', '1971', '1972','1973','1974','1975','1976','1977','1978','1979'])

df_70 = df[df_70]
df_sort = df_70.sort_values(by = ['Total_Sales'],ascending=False)
df_sort
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

df_80 = df['Year'].isin(['1980', '1978', '1982','1983','1984','1985','1986','1987','1988','1989'])

df_80 = df[df_80]

df_sort = df_80.sort_values(by = ['Total_Sales'],ascending=False).head(100)

#df_sort['Publisher'].value_counts()

df_sort
```
